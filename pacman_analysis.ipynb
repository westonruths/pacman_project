```python
def pull_scores(filename) :
    scores = {}
    win_rates = {}
    wins = 0
    with open(filename) as file:
        for line in file:
            s = line.replace(" ", "").strip()
            start = s.find("#") + len("#")
            end = s.find("|s")
            episode = int(s[start:end])

            start = s.find("r:") + len("r:")
            end = s.find("|e")
            score = float(s[start:end])
            scores[episode] = score

            start = s.find("won:") + len("won:")
            end = len(s)
            won = s[start:end]

            if won == "True":
                wins += 1

            win_rates[episode] = wins/episode
    return win_rates
```

```python
#RandomGhost and RandomGhost, 6000, 5000    
win_rates_no_change = pull_scores(filename = "logs/Tue_08_Mar_2022_15_43_18-l-7-m-7-x-5000.log")
```

```python
#RandomGhost and RandomGhostN, 6000, 5000   
win_rates_change = pull_scores("logs/Tue_08_Mar_2022_21_25_11-l-7-m-7-x-5000.log")
```

```python
import matplotlib.pyplot as plt

plt.plot(win_rates_no_change.keys(),win_rates_no_change.values(), label = "No Change in Transition Function")
plt.plot(win_rates_change.keys(),win_rates_change.values(), label = "Change in Transition Function")
plt.title('Win rate plotted against training episodes')
plt.xlabel('Episodes')
plt.ylabel('Win Rate')
plt.legend()
plt.show()
```

```python
win_rates_no_change = pull_scores(filename = "logs/Tue_08_Mar_2022_22_28_21-l-7-m-7-x-5000-RG-RG.log")
win_rates_NS = pull_scores(filename = "logs/Tue_08_Mar_2022_23_01_11-l-7-m-7-x-5000-RG-RGNS.log")
win_rates_EW = pull_scores(filename = "logs/Tue_08_Mar_2022_23_33_13-l-7-m-7-x-5000-RG-RGEW.log")

plt.axvline(x=5000, color='k', linestyle='--')
plt.plot(win_rates_no_change.keys(),win_rates_no_change.values(), label = "No Change (Stochastic then Stochastic)")
plt.plot(win_rates_NS.keys(),win_rates_NS.values(), label = "Change (Stochastic then North/South)")
plt.plot(win_rates_EW.keys(),win_rates_EW.values(), label = "Change (Stochastic then East/West)")
plt.title('Win rate plotted against training episodes \n Comparing Changing Transition Function')
plt.xlabel('Episodes')
plt.ylabel('Win Rate')
plt.legend()
plt.show()
```
