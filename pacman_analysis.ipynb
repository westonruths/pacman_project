<h1> Mid-Term Check-In </h1>

```python
def pull_win_rates(filename) :
    scores = {}
    win_rates = {}
    wins = 0
    with open(filename) as file:
        for line in file:
            s = line.replace(" ", "").strip()
            start = s.find("#") + len("#")
            end = s.find("|s")
            episode = int(s[start:end])

            start = s.find("r:") + len("r:")
            end = s.find("|e")
            score = float(s[start:end])
            scores[episode] = score

            start = s.find("won:") + len("won:")
            end = len(s)
            won = s[start:end]

            if won == "True":
                wins += 1

            win_rates[episode] = wins/episode
    return win_rates
```

```python
def pull_scores(filename) :
    scores = {}
    win_rates = {}
    wins = 0
    total_score = 0
    with open(filename) as file:
        for line in file:
            s = line.replace(" ", "").strip()
            start = s.find("#") + len("#")
            end = s.find("|s")
            episode = int(s[start:end])
            
            start = s.find("r:") + len("r:")
            end = s.find("|e")
            score = float(s[start:end])
            total_score += score
            scores[episode] = total_score

    return scores
```
